```python
from mayavi import mlab
import matplotlib.pyplot as plt
import numpy as np
mlab.init_notebook()
%matplotlib notebook
```

```python
Input='./code/bin/200x2x200/density/15.000299.dat'
nx = 200
ny = 2
nz = 200

data = np.loadtxt("%s"%Input)
density = data[:,0].reshape([nz,nx,ny])
bz = data[:,7].reshape([nz,nx,ny])
bx = data[:,5].reshape([nz,nx,ny])
by = data[:,6].reshape([nz,nx,ny])

#title('lalal')
plt.contourf(by[:,:,0], 40, cmap='RdBu')
plt.colorbar()
```

```python
mlab.pipeline.volume(mlab.pipeline.scalar_field(density),color=(1.,1.,1.))
src = mlab.pipeline.vector_field(bx, by, bz)
s=mlab.pipeline.vectors(src, colormap='plasma',mask_points=20, scale_factor=10., scale_mode='vector', resolution=20, mode='arrow')
#mlab.outline()
s
```
