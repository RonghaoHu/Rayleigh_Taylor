```python
from mayavi import mlab
import matplotlib.pyplot as plt
import numpy as np
mlab.init_notebook()
%matplotlib notebook
```

```python
Input='./code/bin/data/phys_data_30.dat'
nx = 200
ny = 2
nz = 200

data = np.loadtxt("%s"%Input)
density = data[:,0].reshape([nz,nx,ny])
bz = data[:,7].reshape([nz,nx,ny])
bx = data[:,5].reshape([nz,nx,ny])
by = data[:,6].reshape([nz,nx,ny])

#title('lalal')
plt.figure()
plt.subplot(121)
plt.contourf(by[:,:,0], 40, cmap='RdBu')
plt.colorbar()
plt.subplot(122)
plt.contourf(density[:,:,0], 40, cmap='RdBu')
plt.colorbar()
```
